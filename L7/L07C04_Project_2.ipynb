```python
company = ['SOGO']
tech_df_sogo = pdr.get_data_yahoo(company, start = start, end=end)['Adj Close'].pct_change()
company = ['BILI']
tech_df_bili = pdr.get_data_yahoo(company, start = start,end=end)['Adj Close'].pct_change()
company = ['DAO']
tech_df_dao = pdr.get_data_yahoo(company, start = start,end=end)['Adj Close'].pct_change()
```

```python
tech_df_sogo['Name'] = ['SOGO']*len(tech_df_sogo)
tech_df_bili['Name'] = ['BILI']*len(tech_df_bili)
tech_df_dao['Name'] = ['DAO']*len(tech_df_dao)
```

```python
tech_df_sogo = tech_df_sogo.rename(columns = {'SOGO':'Pct'})
tech_df_bili = tech_df_bili.rename(columns = {'BILI':'Pct'})
tech_df_dao = tech_df_dao.rename(columns = {'DAO':'Pct'})
```

```python
tech_dd = pd.concat([tech_df_sogo, tech_df_bili, tech_df_dao], sort=False)
```

```python
tech_dd['DateValue'] = tech_dd.index
```

```python
def getdays(thedate):
    s = thedate - datetime(2019,12,31);
    return s.days
    #return str(s).replace(' days', '')
```

```python
tech_dd['DateValue'] = tech_dd['DateValue'].map(getdays)
```

```python
tech_dd['days from 2020.01.01'] = tech_dd['DateValue']
```

```python
tech_dd = tech_dd.dropna()
```

```python
tech_dd
```

```python
sns.lmplot('days from 2020.01.01', 'Pct',tech_dd,hue='Name', fit_reg = False)
```
