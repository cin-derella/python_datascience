```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(10)
y = 3*x
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,100,3)
y = 3*x*x
x2 = np.arange(1,100,3)
y2 = 5*x*x
plt.plot(x,y,label = 'First Line')
plt.plot(x2,y2,label = 'Second Line')
plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('hello baby')
plt.legend()
plt.show()
```

```python
x = np.random.randint(0,100,[1,200])
y = np.random.randint(0,100,[1,200])
plt.scatter(x,y,label = 'skitscat',color = 'k',s = 25,marker = 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph \nCheck it out')
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(20),np.arange(20,40),label = 'Example one')
plt.show()
```

```python
population_ages = np.random.randint(0,100,[1,100])
bins = np.arange(0,100,10)
plt.hist(population_ages,bins,histtype = 'bar',rwidth = 0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph \nCheck it out')
plt.legend()
plt.show()
```
