```python
x = [1,2,3,4,5]
y = [2,4,6,8,15]
z = [0,3,6,9,12]
plt.title('maths')
plt.xlabel('x--')
plt.ylabel('y--')
plt.plot(x,y)
plt.subplot(3,1,1)
plt.plot(y,z)
plt.subplot(3,1,2)
plt.plot(x,z)
plt.subplot(3,1,3)
plt.show()
```

```python
x = np.arange(10)
y = 2*x
x2 = np.arange(15)
y2 =3*x2
plt.bar(x,y,align = 'center')
plt.subplot(2,1,1)

plt.bar(x2,y2,color = 'g', align = 'center')
plt.subplot(2,1,2)

plt.title('Bar Gragh')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
