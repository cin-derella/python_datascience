```python
import numpy as np
import pandas as pd
```

```python
path = r'../YCdata'
```

```python
pop = pd.read_csv(path + '/' + 'state-population.csv')
areas = pd.read_csv(path + '/' + 'state-areas.csv')
abbrevs = pd.read_csv(path + '/' + 'state-abbrevs.csv')
```

```python
merged = pd.merge(pop,abbrevs,how = 'outer',left_on = 'state/region',
                  right_on = 'abbreviation')
merged
```

```python
merged = merged.drop('abbreviation',axis = 1)  
merged 
```

```python
merged.isnull().any() #查看存在缺失数据的列
```

```python
merged[merged['population'].isnull()] #查看缺失数据
```

```python
merged.loc[merged['state'].isnull(),'state/region']
#产看state为空时，state/region字段的数据
```

```python
merged.loc[merged['state'].isnull(),'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR','state']= 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA','state']= 'United States'
```

```python
merged.isnull().any() #处理了缺失数据
```

```python
final = pd.merge(merged,areas,on = 'state',how='left')  #加入面积
final
```

```python
final.shape
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()]
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.dropna(inplace = True)  #删除nan
```

```python
final
```

```python
final.isnull().any()  #数据都不为空
```

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010
```

```python
data2010.shape
```

```python
data2010.set_index ('state',inplace = True)  # 重置索引
data2010
```

```python
#计算人口密度
density = data2010['population'].div(data2010['area (sq. mi)'])
density
```

```python
density.sort_values(ascending = False, inplace = True) 
#排序，降序，inplace = True本地排序替换density
density.head(10)
```
