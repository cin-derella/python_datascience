```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'age':[18,20,22,22,23,23],
                   'name':['a','b','c','d','e','f'],
                   'price1':np.random.rand(6),
                   'price2':np.random.random(6)})
df
```

```python
result = df.groupby('age')['price1','price2'].\
    apply(sum).add_prefix('a_')
result
```

```python
result = df.groupby('age')['price1','price2'].\
    apply(max).add_prefix('a_')
result
```

```python
result = df.groupby('age')['price1','price2'].\
    transform(max).add_prefix('a_')
result # transform按照自己的索引，非聚合用transform，聚合用apply
```

```python
result = df.groupby('age')['price1','price2'].\
    transform(lambda x : x.sum()).add_prefix('a_')
result 
# transform按照自己的索引，非聚合用transform，聚合用apply,  transform可以自定义函数
```
