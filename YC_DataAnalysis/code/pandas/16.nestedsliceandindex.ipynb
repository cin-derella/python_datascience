```python
import numpy as np
import pandas as pd
```

```python
index = [('beijing',2007),
         ('beijing',2017),
         ('beijing',2027),
         ('shanghai',2007),
         ('shanghai',2017),
         ('shanghai',2027)]

money = np.random.rand(6)
mindex = pd.MultiIndex.from_tuples(index)
city = pd.DataFrame(money,index = mindex)
city
```

```python
index = [(2007,'beijing'),(2007,'shanghai'),
         (2017,'beijing'),(2017,'shanghai'),
         (2027,'beijing'),(2027,'shanghai')]


mindex = pd.MultiIndex.from_tuples(index)
city.reindex(mindex) #重构多层索引
```
