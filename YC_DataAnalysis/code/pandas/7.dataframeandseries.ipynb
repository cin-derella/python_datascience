```python
import numpy as np
import pandas as pd
```

```python
a  = pd.DataFrame(np.random.randint(0,20,(2,2)),
                  columns = list('ab'))
a  # 索引一致才能计算
```

```python
b  = pd.DataFrame(np.random.randint(0,20,(2,2)),columns = list('ab'))
b
```

```python
a+b
```

```python
a  = pd.DataFrame(np.random.randint(0,20,(2,2)),columns = list('ab'))
b  = pd.DataFrame(np.random.randint(0,20,(2,2)),columns = list('bc'))
a+b # 有一方索引不存在，nan
```

```python
a.add(b,fill_value = 0)
```

```python
x = np.random.randint(10,size=(3,4))
x
```

```python
y= np.random.randint(10,size=(4))
y
```

```python
x-y  # 列等计算
```

```python
df = pd.DataFrame(x,columns = list('abcd'))
df
```

```python
df.iloc[0]
```

```python
df.iloc[0,::2] #间隔2
```

```python
df - df.iloc[0,::2]  #不匹配，没有的数据用nan填充
```

```python
df-df.iloc[0] #匹配，每个进行减法
```

```python
df.sub(df.iloc[0,::2],axis = 1) # 1代表行
```

```python
df.sub(df.iloc[0,::2],axis = 0) 
```

```python
df
```

```python
df.sub(pd.Series([100,200,300]),axis = 0) 
```
