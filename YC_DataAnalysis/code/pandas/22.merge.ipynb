```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'name':['huangjian','huanzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','huanzehe','huangjian'],
                    'tall':[170,158,180]})
display(df1,df2)
```

```python
pd.merge(df1,df2) # 一对一归并
```

```python
df1 = pd.DataFrame({'name':['huangjian','huanzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','huanjianjian','huangjian'],
                    'tall':[170,158,180]})
display(df1,df2,pd.merge(df1,df2)) #仅归并匹配的
```

```python
df1 = pd.DataFrame({'name':['huangjian','huanzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','huanzehe','huangjian'],
                    'tall':[170,158,180]})
df3 = pd.DataFrame({'weight':[100,110,120],
                    'tall':[170,158,180]})
display(pd.merge(pd.merge(df1,df2),df3)) # 一一匹配，不能直接df1 df2 df3
```

```python
df1 = pd.DataFrame({'name':['huangjian','huanzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','huanzehe','huangjian'],
                    'tall':[170,158,180]})
df3 = pd.DataFrame({'weight':[100,110,120],
                    'tall':[170,158,1800]})
display(pd.merge(pd.merge(df1,df2),df3)) # 一一匹配，不能直接df1 df2 df3
```

```python
df1 = pd.DataFrame({'name':['huangjian','huanzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','huanzehe','huangjian'],
                    'tall':[170,158,180]})
df3 = pd.DataFrame({'weight':[100,110,120],
                    'tall':[170,158,180]})
df4 = pd.DataFrame({'weight':[100,110,120],
                    'degree':['本科','大专','硕士']})

pd.merge(df1,df2)
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(pd.merge(df1,df2),pd.merge(df3,df4))
```
