```python
columns = ['math','english','chinese','bio','history']
index = ['top2','c9','985','211','1本','2本','3本','大专']
data= {'math':[145,140,135,130,125,120,115,100],
       'english':[145,140,135,130,125,120,115,100],
       'chinese':[145,140,135,130,125,120,115,100],
       'bio':[145*2,140*2,135*2,130*2,125*2,120*2,115*2,100*2]}
df = pd.DataFrame(data = data,index = index,columns = columns)
df  #数字索引
```

```python
# 单行，单列，单个元素，多行，多列，多行多列
# 索引是数字，或自定义
```

```python
df.loc['985']  #单行
```

```python
# df.iloc['985']   报错,iloc[]支持取自定义索引内容，
df.iloc[2]
```

```python
df['math']  # 单列
```

```python
df.math
```

```python
df['math']['211'] #取单个元素
```

```python
df.math['211']
```

```python
df.loc['211']['math']  # loc[]先取行，再取列，211与math换位报错
```

```python
df.iloc[3][0]  #iloc只能按照数字索引取
```

```python
df[['math','chinese']]  # 取出多列
```

```python
# df[['211','985']]  取自定义索引的多行会报错，取数字索引多行ok

df.loc[['211','985']]  # loc[0:2]会报错
```

```python
df.iloc[0:2]
#df.iloc[['211','985']] 会报错
```

```python
df[['math','chinese']][0:3]  #取多列多行
```

```python
df[['math','chinese']].loc[['211','985']]
```

```python
df[['math','chinese']].iloc[0:2]
```
