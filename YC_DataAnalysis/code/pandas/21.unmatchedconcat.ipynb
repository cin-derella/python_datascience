```python
import numpy as np
import pandas as pd
```

```python
def make_dataFrame(cols,ind):
    data = {c:[str(c)+str(i) for i in ind] for c in cols}
    return pd.DataFrame(data,ind)
```

```python
x = make_dataFrame('abc',[1,2])
y = make_dataFrame('bcd',[3,4])
display(x,y,pd.concat([x,y]))  #外链接，维度不一致，得到abcd的并集，与inde无关
```

```python
x = make_dataFrame('abc',[1,2])
y = make_dataFrame('bcd',[2,3])
display(x,y,pd.concat([x,y]))
```

```python
display(x,y,pd.concat([x,y],join = 'inner')) # 内连接，只拼接交集部分
```

```python
display(x,y,pd.concat([x,y],join_axes = [x.columns]))  # 以x为准，x没有的不显示
```

```python
x = make_dataFrame('ab',[0,1])
y = make_dataFrame('ab',[2,3])
display(x,y,pd.concat([x,y]))
```
