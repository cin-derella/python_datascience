```python
df = pd.DataFrame({'item':['ball','mug','pen'],
                'color':['white','rosso','verde']
                })
df
```

```python
price = {
    'ball':5.56,
    'mug':4.20,
    'pen':1.30
}
price['ball']
```

```python
df['price'] = df['item'].map(price)
df  #新建一个series，插入，存在就替换覆盖，不存在就插入
```

```python
data = pd.DataFrame({'zs':[129,130,34],
                     'ls':[136,98,8]})
data
```

```python
def mapscore(score):
    if score < 60:
        return 'Fail'
    elif score >120:
        return 'Awesome'
    else:
        return 'so so'

```

```python
data['ok'] = data['zs'].map(mapscore)
data
```

```python
def mapscore(score):
    return 1

data['times'] = data['zs'].map(mapscore)
data
```

```python
def mapscore(score):
    return str(score)

data['last'] = data['zs'].map(mapscore) 
# map只支持对一维数据进行操作，series可以，dataframe不可以
data
```
