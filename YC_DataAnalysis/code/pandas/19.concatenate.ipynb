```python
import numpy as np
import pandas as pd
```

```python
x = np.array([1,2,3]).reshape(1,3)  #调整形状为二维数组
y = np.array([4,5,6]).reshape(1,3)
z = np.array([7,8,9]).reshape(1,3)
print(x,y,z)
```

```python
np.concatenate([x,y,z])  # numy二维数组的拼接
```

```python
a = np.random.rand(4,4)
a
```

```python
np.concatenate([a,a]) #行拼接 axis =0
```

```python
np.concatenate([a,a],axis=1) #列拼接 axis =1
```

```python
def make_dataFrame(cols,ind):
    data = {c:[str(c)+str(i) for i in ind] for c in cols}
    return pd.DataFrame(data,ind)
```

```python
make_dataFrame('abc',range(3))
```

```python
df1 = make_dataFrame('ab',[1,2])
df2 = make_dataFrame('cd',[1,2])

display(df1,df2,pd.concat([df1,df2],axis =1 ))  # 列拼接,需要保持一致，否则取并集用nan填充
```

```python
df1 = make_dataFrame('ab',[1,2])
df2 = make_dataFrame('cd',[1,3])

display(df1,df2,pd.concat([df1,df2],axis =1 )) 
```

```python
df1 = make_dataFrame('ab',[1,2])
df2 = make_dataFrame('cd',[2,3])
display(df1,df2,pd.concat([df1,df2])) 
```

```python
pd.concat([df1,df2],ignore_index = True) # 覆盖原有索引，重建索引
```
