```python
df = pd.DataFrame(np.random.rand(4,2),
                  columns = ['a','b'],
                  index = [['1','1','2','2'],['a','b','a','b']])
df
```

```python
mindex = pd.MultiIndex.from_arrays([['a','a','b','b'],[1,2,1,2]])
mindex # 双层索引对象，根据list构建
```

```python
df= pd.DataFrame(np.random.rand(4,2),
                columns = ['a','b'],
                index = mindex)
df
```

```python
mindex = pd.MultiIndex.from_tuples([('a',1),('a',2),('b',1),('b',2)])
mindex # 双层索引对象，根据tuple构建

df= pd.DataFrame(np.random.rand(4,2),
                columns = ['a','b'],
                index = mindex)
df
```

```python
mindex = pd.MultiIndex.from_product([['a','b'],[1,2]])
mindex # 双层索引对象，根据product构建

df= pd.DataFrame(np.random.rand(4,2),
                columns = ['a','b'],
                index = mindex)
df
```

```python
mindex = pd.MultiIndex(levels = [['a','b'],[1,2]],
                       labels = [[0,0,1,1],[0,1,0,1]])
mindex # 双层索引对象，根据product构建

df= pd.DataFrame(np.random.rand(4,2),
                columns = ['a','b'],
                index = mindex)
df
```
