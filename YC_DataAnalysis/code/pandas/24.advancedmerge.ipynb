```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'name':['huangjian','huangzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','huangzehe','huangjian'],
                    'tall':[170,158,180],
                    'age':[130,132,133]})

pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on = 'name')  #指定归并的主键
```

```python
df1 = pd.DataFrame({'name':['huangjian','huangzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','lizehe','lijian'],
                    'tall':[170,158,188]})

pd.merge(df1,df2) #内归并，只显示共有的,交集
```

```python
pd.merge(df1,df2,how = 'outer')  #外归并，没有的填充nan，
```

```python
pd.merge(df1,df2,how= 'left') #以左为准,df2
```

```python
pd.merge(df1,df2,how= 'right') #以右为准，df2
```

```python
df1 = pd.DataFrame({'name':['huangjian','huangzehe','huangtaoqi'],
                    'age':[30,32,33]})
df2 = pd.DataFrame({'name':['huangtaoqi','huangzehe','huangjian'],
                    'age':[170,158,188]})

pd.merge(df1,df2)  # age列冲突
```

```python
pd.merge(df1,df2,on = 'name',suffixes=['_x','_y']) #对age加后缀，解决列冲突
```
