```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
path = '../YCdata/stock.txt'
stock = pd.read_table(path,sep='\t',index_col = "Trddt")
stock.head(5)
```

```python
stock.shape
```

```python
HXBank =stock[stock['Stkcd'] ==600015]
HXBank
```

```python
HXBank.index = pd.to_datetime(HXBank.index)
HXBank
```

```python
HXRet = HXBank.Dretwd
HXRet
```

```python
HXRet.name = 'HXRet'
HXRet
```

```python
HXRet.plot()
```

```python
newPath = '../YCdata/ThreeFactors.txt'
ThreeFactors = pd.read_table(newPath,sep ='\t',index_col='TradingDate')
ThreeFactors
```

```python
ThreeFactors.index = pd.to_datetime(ThreeFactors.index)
ThreeFactors
```

```python
ThreeFac = ThreeFactors['2014-01-02':]
ThreeFac
```

```python
ThreeFac = ThreeFac.iloc[:,[2,4,6]]
ThreeFac
```

```python
LastFac = pd.merge(pd.DataFrame(HXRet),pd.DataFrame(ThreeFac),left_index=True,right_index=True)
LastFac
```

```python
LastFac.plot()
```

```python
plt.subplot(221)
plt.scatter(LastFac.HXRet,LastFac.RiskPremium2)
plt.subplot(222)
plt.scatter(LastFac.HXRet,LastFac.SMB2)
plt.subplot(223)
plt.scatter(LastFac.HXRet,LastFac.HML2)
```

```python
import statsmodels.api as sm
regFac = sm.OLS(LastFac.HXRet,sm.add_constant(LastFac.iloc[:,1:4]))
result = regFac.fit()
result.summary()
```
