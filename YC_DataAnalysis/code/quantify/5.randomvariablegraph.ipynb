```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
```

```python
plt.plot(np.arange(0,5,0.02),stats.chi.pdf(np.arange(0,5,0.02),3))
plt.title('卡方分布')


```

```python
x = np.arange(-4,4,0.004)
x
```

```python
x = np.arange(-4,4.004,0.004)
x
```

```python
plt.plot(x,stats.norm.pdf(x),label = '正态分布')
plt.plot(x,stats.t.pdf(x,5),label = 'T分布')
plt.legend()

plt.plot(x,stats.f.pdf(x,4,40),label = 'F分布')
plt.legend()
```
