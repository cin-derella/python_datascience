```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [10.1,10.12,9.9,10.7,9.97,10.132,10.72,10.18,9.89,9.95]
x
```

```python
np.mean(x)
```

```python
stats.sem(x)  # 样本的平均差
```

```python
stats.t.interval(0.95,len(x)-1,np.mean(x),stats.sem(x))
```

```python
path = r'../YCdata/TRD_Index.csv'
SHindex = pd.read_csv(path)
SHindex
```

```python
Retindex = SHindex['Retindex']
Retindex
```

```python
Retindex.hist()
```

```python
meanData = Retindex.mean()
stdData = Retindex.std()
plt.hist(Retindex,density = True)
plt.plot(np.arange(-0.06,0.062,0.002),
         stats.norm.pdf(np.arange(-0.6,0.062,0.002),meanData,stdData))
```

```python
stats.t.interval(0.85,len(Retindex)-1,meanData,stats.sem(Retindex))
```
