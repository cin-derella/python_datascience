```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
path = '../YCdata/TRD_Index.csv'
indexcd = pd.read_csv(path,index_col = 'Trddt')
indexcd.head()
```

```python
mktcd = indexcd[indexcd.Indexcd ==902]
mktcd
```

```python
mktret = pd.Series(mktcd.Retindex.values,index = pd.to_datetime(mktcd.index))
mktret
```

```python
mktret = mktret['2014-01-02':'2014']
mktret
```

```python
mktret.plot()
```

```python
xinanpath = '../YCdata/xin_an.csv'
xin_an = pd.read_csv(xinanpath,index_col='Date')
xin_an
```

```python
xin_an.index = pd.to_datetime(xin_an.index) #修改索引
xin_an
```

```python
xin_an = xin_an[xin_an.Volume!=0]
xin_an
```

```python
xin_anret = (xin_an.Close-xin_an.Close.shift(1))/xin_an.Close.shift(1)
xin_anret.name = 'xin_anret'
xin_anret = xin_anret.dropna()
xin_anret
```

```python
xin_anret.plot()
```

```python
mktret.name = 'mktret'
Ret = pd.merge(pd.DataFrame(mktret),pd.DataFrame(xin_anret),
               left_index=True,right_index=True,how = 'inner')
Ret
```

```python
Ret.plot()
```

```python
rf = 1.036**(1/360)-1
rf #最小系数
```

```python
Eret = Ret - rf
Eret
```

```python
plt.scatter(Eret.values[:,0],Eret.values[:,1])
plt.title('散点图')
```

```python
import statsmodels.api as sm
model = sm.OLS(Eret.mktret[1:],sm.add_constant(Eret.mktret[1:]))
result = model.fit()
result.summary()
```

```python
import statsmodels.api as sm
model = sm.OLS(Eret.xin_anret[1:],sm.add_constant(Eret.xin_anret[1:]))
result = model.fit()
result.summary()
```
