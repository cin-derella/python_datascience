```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
path1 = r'../YCdata/TRD_Year.csv'
year_return = pd.read_csv(path1,encoding='gbk')
year_return
```

```python
import statsmodels.stats.anova as anova
from statsmodels.formula.api import ols
```

```python
model = ols('Return ~ (Industry)',\
            data = year_return.dropna()).fit() #建立模型
table = anova.anova_lm(model) #根据模型建立表格
table
```

```python
path2 = r'../YCdata/PSID.csv'
PSID = pd.read_csv(path2,encoding='gbk')
PSID
```

```python
model1 = ols('earnings~ C (married) + C(educatn)',\
            data = PSID.dropna()).fit() #建立模型
table1 = anova.anova_lm(model1) #根据模型建立表格
table1
```

```python
model1 = ols('earnings~ C (married) * C(educatn)',\
            data = PSID.dropna()).fit() #建立模型
table1 = anova.anova_lm(model1) #根据模型建立表格
table1
```
