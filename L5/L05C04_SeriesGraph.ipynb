```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
s1 = Series(np.random.randn(10)).cumsum()
```

```python
s1.plot(kind = 'line',grid = True,label = 'S1',title='this is a series',style = '--')
plt.legend()
```

```python
s1 = Series(np.random.randn(10)).cumsum()
s2 = Series(np.random.randn(10)).cumsum()

s1.plot(kind = 'line',grid = True,label = 'S1',title='this is a series',style = '--')
s2.plot(label = 'S2')

plt.legend()
```

```python
fig,ax = plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
```

```python
fig,ax = plt.subplots(2,1)
s1[0:5].plot(ax = ax[0], kind = 'bar')
s2.plot(ax = ax[1])
```
