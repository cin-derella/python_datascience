```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
```

```python
sinplot()
```
